```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import precision_score, accuracy_score
from sklearn.model_selection import train_test_split

# Load your dataset
df = pd.read_csv("final_rolling.csv")

# Convert date to datetime
df["date"] = pd.to_datetime(df["date"])

# Drop rows with missing values (or use imputation)
df = df.dropna(subset=["dist_rolling"])

# Select features (mix of categorical + numeric)
features = [
    'venue_code', 'opp_code', 'hour', 'day_code',
    'gf_rolling', 'ga_rolling', 'dist_rolling'
]

# Split data into train/test based on date
train = df[df["date"] < "2022-03-01"]
test = df[df["date"] >= "2022-03-01"]

train_X = train[features]
train_y = train["target"]
test_X = test[features]
test_y = test["target"]

# Define grid search parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'max_features': ['sqrt', 'log2']
}

# Grid Search on Random Forest
grid_search = GridSearchCV(
    RandomForestClassifier(class_weight='balanced', random_state=1),
    param_grid,
    scoring='accuracy',
    cv=5,
    n_jobs=-1
)

grid_search.fit(train_X, train_y)

# Best model from grid search
best_rf = grid_search.best_estimator_

# Predict on test data
preds = best_rf.predict(test_X)

# Evaluation
from sklearn.metrics import precision_score

precision = precision_score(test_y, preds, average='macro')  # or 'weighted'

# precision = precision_score(test_y, preds)
accuracy = accuracy_score(test_y, preds)

print("Best Parameters:", grid_search.best_params_)
print("Test Accuracy:", accuracy)
print("Test Precision:", precision)
```
